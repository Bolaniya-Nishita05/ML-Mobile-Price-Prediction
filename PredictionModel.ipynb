```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pickle
```

```python
df=pd.read_csv("Mobile phone price.csv")
```

```python
df
```

```python
df.info()
```

```python
df["Brand"].unique()
```

```python
df["Storage"].unique()
```

```python
df["Storage"]=np.array([int(value.replace('GB', '').replace(' ','').replace(',', '')) for value in df["Storage"]])
```

```python
df["RAM"].unique()
```

```python
df["RAM"]=np.array([int(value.replace('GB', '').replace(' ','').replace(',', '')) for value in df["RAM"]])
```

```python
df["Screen Size (inches)"].unique()
```

```python
df["Screen Size (inches)"] = df["Screen Size (inches)"].str.replace(r"[^0-9.]", "", regex=True).str.replace(r"\.(?=.*\.)", "", regex=True)
```

```python
df[df["Screen Size (inches)"]==683.9] = 6.83
```

```python
df["Screen Size (inches)"] = df["Screen Size (inches)"].astype(float)
```

```python
df["Price ($)"].unique()
```

```python
df["Price ($)"]=np.array([float(value.replace('$', '').replace(',', '')) for value in df["Price ($)"]])
```

```python
import seaborn as sns
sns.heatmap(df[["Storage","RAM","Screen Size (inches)","Battery Capacity (mAh)", "Price ($)"]].corr(), annot=True)
```

```python
df[["Storage","RAM","Screen Size (inches)","Battery Capacity (mAh)", "Price ($)"]].corr()
```

```python
d={}
n=1
print(n)

for i in df["Brand"].unique():
    d[i]=n
    n=n+1
    
d
```

```python
df["Brand"]=df["Brand"].map(d)
```

```python
df
```

```python
sns.heatmap(df[["Brand","Storage","RAM","Screen Size (inches)","Battery Capacity (mAh)", "Price ($)"]].corr(), annot=True)
```

```python
df[["Brand","Storage","RAM","Screen Size (inches)","Battery Capacity (mAh)", "Price ($)"]].corr()
```

```python
x=df[["Brand","Storage","RAM","Screen Size (inches)","Battery Capacity (mAh)"]]
x
```

```python
y=df["Price ($)"]
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7)
```

```python
sc=StandardScaler()
x_train_scaled=sc.fit_transform(x_train)
```

```python
model=RandomForestRegressor(n_estimators=100)
```

```python
model.fit(x_train_scaled,y_train)
```

```python
model.score(sc.transform(x_test),y_test)
```

```python
model.predict(sc.transform(x_test))
```

```python
f=open("PriceModel.pkl","wb")
pickle.dump(model,f)
f.close()
```

```python
f=open("PriceModel.pkl","rb")
model1=pickle.load(f)
f.close()
```

```python
model1.score(sc.transform(x_test),y_test)
```
